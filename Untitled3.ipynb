1. Для каждого из текстов подсчитайте и выведите кол-во символов, кол-во слов и кол-во предложений в текстах. Определите и выведите на экран три самых длинных слова в текстах. Для каждого из текстов подсчитайте и выведите минимальное, максимальное и среднее кол-во слов в предложениях текста.

```python
print("1. БАЗОВАЯ СТАТИСТИКА ТЕКСТОВ")

def basic_statistics(text):
    # Количество символов
    char_count = len(text)

    # Количество слов
    words = text.split()
    word_count = len(words)

    # Количество предложений
    sentences = re.split(r'[.!?]+', text)
    sentences = [s.strip() for s in sentences if s.strip()]
    sentence_count = len(sentences)

    # Статистика по словам в предложениях
    sentence_word_counts = [len(sentence.split()) for sentence in sentences]
    min_words = min(sentence_word_counts) if sentence_word_counts else 0
    max_words = max(sentence_word_counts) if sentence_word_counts else 0
    avg_words = sum(sentence_word_counts) / len(sentence_word_counts) if sentence_word_counts else 0

    # Три самых длинных слова
    clean_words = [re.sub(r'[^\w]', '', word) for word in words]
    longest_words = sorted(set(clean_words), key=len, reverse=True)[:3]

    return {
        'char_count': char_count,
        'word_count': word_count,
        'sentence_count': sentence_count,
        'min_words_per_sentence': min_words,
        'max_words_per_sentence': max_words,
        'avg_words_per_sentence': avg_words,
        'longest_words': longest_words
    }

# Анализ каждого текста
for num, text in texts.items():
    stats = basic_statistics(text)
    print(f"\n--- text_{num}.txt ---")
    print(f"Количество символов: {stats['char_count']}")
    print(f"Количество слов: {stats['word_count']}")
    print(f"Количество предложений: {stats['sentence_count']}")
    print(f"Минимальное количество слов в предложении: {stats['min_words_per_sentence']}")
    print(f"Максимальное количество слов в предложении: {stats['max_words_per_sentence']}")
    print(f"Среднее количество слов в предложении: {stats['avg_words_per_sentence']:.2f}")
    print(f"Три самых длинных слова: {stats['longest_words']}")
```

2. Оцените и выведите, какой из текстов более читабелен по критерию индекса лёгкости чтения по шкале Флеша и индексу туманности Ганнинга.

```python
print("2. ОЦЕНКА ЧИТАБЕЛЬНОСТИ")


def calculate_readability(text):
    try:
        readability = Textatistic(text)
        scores = readability.scores
        return {
            'flesch': scores['flesch_score'],
            'gunning_fog': scores['gunningfog_score']
        }
    except Exception as e:
        print(f"Ошибка при расчете читабельности: {e}")
        return None

readability_results = {}
for num, text in texts.items():
    result = calculate_readability(text)
    if result:
        readability_results[num] = result
        print(f"\n--- text_{num}.txt ---")
        print(f"Индекс лёгкости чтения Флеша: {result['flesch']:.2f}")
        print(f"Индекс туманности Ганнинга: {result['gunning_fog']:.2f}")

# Определение наиболее читабельного текста
if readability_results:
    best_flesch = max(readability_results.items(), key=lambda x: x[1]['flesch'])
    best_gunning = min(readability_results.items(), key=lambda x: x[1]['gunning_fog'])

    print(f"\nСамый читабельный текст по Флешу: text_{best_flesch[0]}.txt ({best_flesch[1]['flesch']:.2f})")
    print(f"Самый читабельный текст по Ганнингу: text_{best_gunning[0]}.txt ({best_gunning[1]['gunning_fog']:.2f})")
```

3. Для каждого из текстов выполните токенизацию и определите пять наиболее часто встречающихся токена и частоты их повторения.

```python
print("3. ТОКЕНИЗАЦИЯ И ЧАСТОТНЫЙ АНАЛИЗ")

def tokenize_and_analyze(text):
    doc = nlp(text)

    # Токенизация
    tokens = [token.text.lower() for token in doc if not token.is_space and not token.is_punct]

    # Частотный анализ
    token_freq = Counter(tokens)

    return token_freq

for num, text in texts.items():
    freq_analysis = tokenize_and_analyze(text)
    print(f"\n--- text_{num}.txt ---")
    print("5 наиболее частых токенов:")
    for token, freq in freq_analysis.most_common(5):
        print(f"  {token}: {freq}")
```

6. Для каждого из текстов выполните разметку частей речи, определите и выведите кол-во токенов каждого типа. Визуализируйте кол-во токенов каждого типа в текстах в виде столбчатых диаграмм.

```python
print("6. РАЗМЕТКА ЧАСТЕЙ РЕЧИ (POS TAGGING)")


def pos_analysis(text):
    doc = nlp(text)

    # Подсчет POS-тегов
    pos_tags = [token.pos_ for token in doc if not token.is_space]
    pos_freq = Counter(pos_tags)

    return pos_freq

pos_counts = {}
for num, text in texts.items():
    pos_freq = pos_analysis(text)
    pos_counts[num] = pos_freq

    print(f"\n--- text_{num}.txt ---")
    for pos, count in pos_freq.most_common():
        print(f"  {pos}: {count}")

# Визуализация POS-тегов
plt.figure(figsize=(15, 5))
for i, num in enumerate(texts.keys(), 1):
    plt.subplot(1, len(texts), i)
    pos_freq = pos_counts[num]

    # Берем топ-10 самых частых POS-тегов
    top_pos = dict(pos_freq.most_common(10))

    plt.bar(top_pos.keys(), top_pos.values(), color=f'C{i-1}')
    plt.title(f'text_{num}.txt - POS теги')
    plt.xlabel('Тип POS')
    plt.ylabel('Количество')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

7. Для каждого из текстов найдите именованные сущности, выведите и подсчитайте их кол-во для каждого типа.

```python
print("7. РАСПОЗНАВАНИЕ ИМЕНОВАННЫХ СУЩНОСТЕЙ (NER)")


def named_entity_recognition(text):
    doc = nlp(text)

    entities = {}
    for ent in doc.ents:
        if ent.label_ not in entities:
            entities[ent.label_] = []
        entities[ent.label_].append(ent.text)

    return entities

for num, text in texts.items():
    entities = named_entity_recognition(text)
    print(f"\n--- text_{num}.txt ---")
    for label, ents in entities.items():
        print(f"  {label}: {len(ents)} сущностей")
        # Показываем первые 3 сущности каждого типа
        for ent in ents[:3]:
            print(f"    - {ent}")
```

8. Для имеющихся текстов постройте векторы tf-idf и выведите форму (shape) построенной матрицы.

```python
print("8. TF-IDF ВЕКТОРИЗАЦИЯ")


# Создаем корпус из всех текстов
corpus = list(texts.values())

# Создаем TF-IDF векторы
vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)
tfidf_matrix = vectorizer.fit_transform(corpus)

print(f"Форма TF-IDF матрицы: {tfidf_matrix.shape}")
print(f"Количество документов: {tfidf_matrix.shape[0]}")
print(f"Количество признаков (слов): {tfidf_matrix.shape[1]}")

# Показываем топ-10 самых важных слов для каждого документа
feature_names = vectorizer.get_feature_names_out()
for i, num in enumerate(texts.keys()):
    print(f"\n--- text_{num}.txt ---")
    # Получаем TF-IDF значения для текущего документа
    feature_index = tfidf_matrix[i,:].nonzero()[1]
    tfidf_scores = zip(feature_index, [tfidf_matrix[i, x] for x in feature_index])

    # Сортируем по убыванию TF-IDF
    sorted_scores = sorted(tfidf_scores, key=lambda x: x[1], reverse=True)

    print("Топ-10 самых важных слов:")
    for idx, score in sorted_scores[:10]:
        print(f"  {feature_names[idx]}: {score:.4f}")
```
